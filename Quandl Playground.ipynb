```python
import random
import numpy as np
import pandas as pd
point1 =np.random.randint(0,2,(12000,13))
```

```python
df=pd.DataFrame(point1)
df.columns = ['Is Military','Is Civilian','Flew Fighter','Flew Transport','Flew Corporate','Flew RJ',
              'Arts Degree','Stem Degree','A-Hours','B-Hours','C-Hours','Prev Capt','Prev Rot-Wing']
```

```python
import seaborn as sns
from scipy import stats
from scipy.stats import norm, skewnorm

#choosing parameters for our skewed distribution
a, loc, scale = 5,1,1

#create 280,000 random variables for our distribution
df["# of retrains"] = stats.skewnorm(a, loc, scale).rvs(12000)
```

```python
from matplotlib import pyplot as plt
#plotting data set #2
fig, ax = plt.subplots(figsize = (12,8))
sns.histplot(df["# of retrains"], bins=5, color = 'green', kde = True)
legends = ['##']
plt.legend(legends)
ax.set_title("Histogram of retrains")
ax.set_xlabel("# of Retrains")
ax.set_ylabel("Frequency")
plt.axvline(np.mean(df["# of retrains"]),linewidth = 4)
```

```python
X=df
y=df["# of retrains"]
```

```python
len(X)
```

```python
from sklearn.ensemble import GradientBoostingRegressor as gbr
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,y,random_state=42)
```

```python
estimator = gbr()
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.score(X_test, y_test)
```
